### Standard RPA calcualtion for CO

```
memory,4000,m  
gthresh,twoint=1d-20,energy=1d-10,orbital=1d-8  
basis={  
default,aug-cc-pwCVQZ  
set,ri
default,aug-cc-pwCVQZ/mp2fit  
}

symmetry,nosym  
angstrom  
geometry={  
2  

C        0.000000    0.000000   -0.646514 
O        0.000000    0.000000    0.484886  
}  

charge=0  

spin=0  

grid,name=neese,neese_index=6  

df-ks,pbe,df_basis=aug-cc-pwCV5Z/mp2fit,maxit=200  
{cfit,basis_coul=aug-cc-pwCV5Z/mp2fit,basis_exch=aug-cc-pwCV5Z/mp2fit}  

user,rirpa,orb=2100.2,nquad=50,rpa=1,\  
charge=1,homo=0,\  
dfit=1,direct=1,\  
thr_overlap_ri=1d-10,thr_fai_ri=1d-10  
```

This gives:

correlation energy =         -0.802657418904

### Sigma-scaled RPA calcualtion for CO
Note that Nodefile is required for this calculation

```
memory,4000,m  
gthresh,twoint=1d-20,energy=1d-10,orbital=1d-8  
basis={  
default,aug-cc-pwCVQZ  
set,ri
default,aug-cc-pwCVQZ/mp2fit  
}

symmetry,nosym  
angstrom  
geometry={  
2  

C        0.000000    0.000000   -0.646514 
O        0.000000    0.000000    0.484886  
}  

charge=0  

spin=0  

grid,name=neese,neese_index=6  

df-ks,pbe,df_basis=aug-cc-pwCV5Z/mp2fit,maxit=200  
{cfit,basis_coul=aug-cc-pwCV5Z/mp2fit,basis_exch=aug-cc-pwCV5Z/mp2fit}  

user,rirpa,orb=2100.2,nquad=50,rpa=1,\  
charge=1,homo=0,\  
dfit=1,direct=1,\  
thr_overlap_ri=1d-10,thr_fai_ri=1d-10,\  
sigma_scale=1,\
vc_scal=1d0  
```

This gives:

correlation energy =         -0.694102192073

### Standard RPA calcualtion for Al

```
memory,4000,m  
gthresh,twoint=1d-20,energy=1d-10,orbital=1d-8  
basis={  
default,aug-cc-pwCVQZ  
set,ri  
default,aug-cc-pwCVQZ/mp2fit  
}

symmetry,nosym  
angstrom  
geometry={  
1  

Al 0.0 0.0 0.0  
}  

charge=0  

spin=1  

grid,name=neese,neese_index=6  

df-uks,pbe,df_basis=aug-cc-pwCV5Z/mp2fit,maxit=200  
{cfit,basis_coul=aug-cc-pwCV5Z/mp2fit,basis_exch=aug-cc-pwCV5Z/mp2fit}  

user,urirpa,orb=2200.2,nquad=50,rpa=1,\  
charge=3,homo=0,\  
dfit=1,direct=1,\  
thr_overlap_ri=1d-10,thr_fai_ri=1d-10 
```

This gives:

correlation energy =         -0.622966103914

### Sigma-scaled RPA calcualtion for Al
Note that Nodefile is required for this calculation

```
memory,4000,m  
gthresh,twoint=1d-20,energy=1d-10,orbital=1d-8  
basis={  
default,aug-cc-pwCVQZ  
set,ri  
default,aug-cc-pwCVQZ/mp2fit  
}

symmetry,nosym  
angstrom  
geometry={  
1  

Al 0.0 0.0 0.0  
}  

charge=0  

spin=1  

grid,name=neese,neese_index=6  

df-uks,pbe,df_basis=aug-cc-pwCV5Z/mp2fit,maxit=200  
{cfit,basis_coul=aug-cc-pwCV5Z/mp2fit,basis_exch=aug-cc-pwCV5Z/mp2fit}  

user,urirpa,orb=2200.2,nquad=50,rpa=1,\  
urirpa_mode=3,\ ! this is required to use sigma-scaling in spin-polarized calculations  
charge=3,homo=0,\  
dfit=1,direct=1,\  
thr_overlap_ri=1d-10,thr_fai_ri=1d-10,\  
sigma_scale=1,\  
vc_scal=1d0
```

This gives:

correlation energy =         -0.546128215459
